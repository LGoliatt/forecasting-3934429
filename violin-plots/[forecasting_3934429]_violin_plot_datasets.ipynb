```python
# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import matplotlib.dates as mdates

# Miscellaneous
import pandas as pd
import numpy as np
import math
import joblib
import gdown
```

# Dataset D1

```python
# File ID in Google Drive
# https://drive.google.com/file/d/1ZJIN2zmzcJeP989FZc8pZo48q7aD7Htr/view?usp=sharing
file_id = "1ZJIN2zmzcJeP989FZc8pZo48q7aD7Htr"

# Correct URL for download
download_url = f"https://drive.google.com/uc?id={file_id}"

# File name after download
output_file = "d1.csv"

# Download the file
gdown.download(download_url, output_file, quiet=False)

# Click on Pandas
df1 = pd.read_csv(output_file)

# Checking
df1.head()
```

```python
# Create violin plots for each column in the dataframe
for i, column in enumerate(df1.columns):
    plt.figure(figsize=(8, 6))
    sns.violinplot(y=df1[column], color=sns.color_palette('Set1', len(df1.columns))[i])
    plt.title(f'Violin Plot of {column}')
    plt.ylabel(column)
    plt.show()
```
